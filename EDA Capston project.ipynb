```python
#15. Create a joint point with body_length_mm, body_depth_mm for penguins_cleaned dataset
df.isnull().sum()
```

```python
df["bill_length_mm"].fillna(value = 0)
df["bill_length_mm"].fillna(value = "missing")
df["bill_length_mm"] = df["bill_length_mm"].fillna(value = df["bill_length_mm"].mean())
df["bill_length_mm"].fillna(value = df["bill_length_mm"].mode())
(df.isnull().sum())
```

```python
df["bill_depth_mm"].fillna(value = 0)
df["bill_depth_mm"].fillna(value = "missing")
df["bill_depth_mm"] = df["bill_depth_mm"].fillna(value = df["bill_depth_mm"].mean())
df["bill_depth_mm"].fillna(value = df["bill_depth_mm"].mode())
(df.isnull().sum())
```

```python
df["flipper_length_mm"].fillna(value = 0)
df["flipper_length_mm"].fillna(value = "missing")
df["flipper_length_mm"] = df["flipper_length_mm"].fillna(value = df["flipper_length_mm"].mean())
df["flipper_length_mm"].fillna(value = df["flipper_length_mm"].mode())
(df.isnull().sum())
```

```python
df["body_mass_g"].fillna(value = 0)
df["body_mass_g"].fillna(value = "missing")
df["body_mass_g"] = df["body_mass_g"].fillna(value = df["body_mass_g"].mean())
df["body_mass_g"].fillna(value = df["body_mass_g"].mode())
(df.isnull().sum())
```

```python
df["sex"].fillna(value = 0)
df["sex"].fillna(value = "missing")
df["sex"].fillna(value = df["sex"].mode())
(df.isnull().sum())
```

```python
sns.jointplot(x = "bill_length_mm", y = "bill_depth_mm", data = df)
plt.show()
```
